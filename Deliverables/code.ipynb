```python
import pandas as pd
```

```python
#Reading the csv files

appearances = pd.read_csv('archive/appearances.csv')
club_games = pd.read_csv('archive/club_games.csv')
competitions = pd.read_csv('archive/competitions.csv')
game_events = pd.read_csv('archive/game_events.csv')
game_lineups = pd.read_csv('archive/game_lineups.csv')
games = pd.read_csv('archive/games.csv')
player_valuations = pd.read_csv('archive/player_valuations.csv')
players = pd.read_csv('archive/players.csv')
clubs = pd.read_csv('archive/clubs.csv')
```

```python
#Filtering for active players in players

condition_active_players = (players['last_season'] == 2023)
active_players = players[condition_active_players]
active_players.head()
```

```python
game_lineups.info()
```

```python
game_lineups.head()
```

```python
#Filtering for active players in game_lineups

active_players_id = active_players["player_id"]
active_game_lineups = game_lineups[game_lineups["player_id"].isin(active_players_id)]
active_game_lineups.head()
```

```python
#Filtering for active players in player valuations

active_player_valuations = player_valuations[player_valuations["player_id"].isin(active_players_id)]
active_player_valuations.info()
```

```python
#Filtering for active players in games using active_game_lineups

active_games_id = active_game_lineups["game_id"]
active_games = games[games["game_id"].isin(active_games_id)]
active_games.info()
```

```python
#Filtering for active players in game_events using active_game_lineups

active_games_id = active_game_lineups["game_id"]
active_game_events = game_events[game_events["game_id"].isin(active_games_id)]
active_game_events.info()
```

```python
#Filtering for active players in club_games using active_game_lineups

active_games_id = active_game_lineups["game_id"]
active_club_games = club_games[club_games["game_id"].isin(active_games_id)]
active_club_games.info()
```

```python
#Filtering for active players in appearances using active_game_lineups

active_games_id = active_game_lineups["game_id"]
active_appearances = appearances[appearances["game_id"].isin(active_games_id)]
active_appearances.info()
```
