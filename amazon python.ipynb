Data Cleaning

Drop Columns

```python
amazon.drop(columns= ['Unnamed: 22', 'fulfilled-by', 'ship-country', 'currency', 'Sales Channel '], inplace = True)
```

Drop duplicates, if there is

```python
amazon[amazon.duplicated(['Order ID','ASIN'], keep=False)]
```

```python
#check how many duplicates would be dropped dropped
len(amazon)-len(amazon.drop_duplicates(['Order ID','ASIN']))
```

```python
amazon.drop_duplicates(['Order ID','ASIN'],inplace = True,ignore_index=True)
```

Replacing NaN values

```python
amazon['Courier Status'].fillna('unknown',inplace=True)
```

```python
amazon['promotion-ids'].fillna('no promotion',inplace=True)
```

```python
# distribution of statuses for the entries where the amount is null.

amazon[amazon['Amount'].isnull()]['Status'].value_counts(normalize=True).apply(lambda x: format(x, '.2%'))
```

```python
# replace NaN in amount by 0

amazon['Amount'].fillna(0,inplace=True)
```

```python
# replace NaN by 0
amazon['ship-city'].fillna('unknown', inplace = True)
amazon['ship-state'].fillna('unknown', inplace = True)
amazon['ship-postal-code'].fillna('unknown', inplace = True)
```

Renaming Columns

```python
mapper = {'Order ID':'order_ID', 'Date':'date', 'Status':'ship_status','Fulfilment':'fullfilment',
          'ship-service-level':'service_level', 'Style':'style', 'SKU':'sku', 'Category':'product_category', 
          'Size':'size', 'ASIN':'asin', 'Courier Status':'courier_ship_status', 'Qty':'order_quantity', 
          'Amount':'order_amount_($)', 'ship-city':'city', 'ship-state':'state', 'ship-postal-code':'zip', 
          'promotion-ids':'promotion','B2B':'customer_type'}
```

```python
amazon.rename(columns=mapper,inplace=True)
```

Convert amount values to USD. INR = 0.012

```python
exchange_rate = 0.012

amazon['order_amount_($)'] = amazon['order_amount_($)'].apply(lambda x: x * exchange_rate)
```

Make B2B column more readable

```python
amazon['customer_type'].replace(to_replace=[True,False],value=['business','customer'],inplace=True)
```

Convert date to Datetime and creating Month column

```python
amazon['date'] = pd.to_datetime(amazon['date'])
```

```python
# filter to only include data in March

march_dates = amazon['date'][amazon['date'].dt.month == 3]

# number of unique days in March
march_dates.dt.day.nunique()
```

```python
# dropping March dates from the dataset
amazon = amazon[(amazon['date'].dt.month != 3)]
```

```python
# making sure we only have 3 months
amazon['month'] = amazon['date'].dt.month
amazon["month"].unique()
```

```python
month_map = { 4: 'april',5: 'may',6: 'june'}
amazon['month'] = amazon['date'].dt.month.map(month_map)
```

```python
# Define the desired order of months
month_order = ['april', 'may', 'june']

# Convert the month column to a categorical data type with the desired order
amazon['month'] = pd.Categorical(amazon['month'], categories=month_order, ordered=True)
```

```python
print(f'This dataset contains the months {amazon["month"].unique()} for 2022')
print(f'The earliest date is {amazon["date"].min()}')
print(f'The latest date is {amazon["date"].max()}')
```

Order Data

Business questions (to be also answered in SQL)

*What is the revenue by month?*

```python
revenue_by_month = amazon.groupby('month')['order_amount_($)'].sum()
revenue_by_month
```

*What is the decrease MoM?*

```python
percent_decrease_apr_to_may = (revenue_by_month['april'] - revenue_by_month['may']) / revenue_by_month['april'] * 100
percent_decrease_may_to_jun = (revenue_by_month['may'] - revenue_by_month['june']) / revenue_by_month['may'] * 100
total_decrease = (revenue_by_month['april'] - revenue_by_month['june']) / revenue_by_month['april'] * 100

print(f'The percent of decrease from April to May is {percent_decrease_apr_to_may:.2f}%')
print(f'The percent of decrease from May to Jun is {percent_decrease_may_to_jun:.2f}%')
print(f'The total percent of decrease is {total_decrease:.2f}%')
```

```python
share_revenue_category = (revenue_by_category/sum(revenue_by_category)*100).sort_values(ascending=False)
share_revenue_category
```

*How many orders got cancelled?*

```python
cancelled_status = ['Cancelled', 'Shipped - Lost in Transit']
total_cancelled = len(amazon[amazon['ship_status'].isin(cancelled_status)])
print(f'Total orders cancelled: {total_cancelled}')

percent_cancelled = total_cancelled/len(amazon)*100
print(f'Percentage of orders cancelled: {percent_cancelled:.2f}%')
```

*How many orders got returned?*

```python
returned_status = ['Shipped - Returned to Seller', 'Shipped - Returning to Seller', 'Shipped - Rejected by Buyer', 'Shipped - Damaged']
total_returned = len(amazon[amazon['ship_status'].isin(returned_status)])
print(f'Total orders returned: {total_returned}')

percent_returned = total_returned/len(amazon)*100
print(f'Percentage of orders returned: {percent_returned:.2f}%')
```

*What are the most popular category by state?*

```python
most_popular_products_state = amazon.groupby(['state', 'product_category'])['order_quantity'].sum().reset_index()
most_popular_products_state = most_popular_products_state.sort_values(['state', 'order_quantity'], ascending=[True, False])
most_popular_products_state = most_popular_products_state.drop_duplicates(subset=['state'])

most_popular_products_state
```

*What is the AOV by customer type?*

```python
# AOV = avg(amount)

aov_customer_type = amazon.groupby('customer_type')['order_amount_($)'].mean()
print(f'Avg. Order by Customer type: \n {aov_customer_type}')
```
