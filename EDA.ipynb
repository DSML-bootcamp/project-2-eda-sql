```python
df['CustomerID']= df['CustomerID'].astype(int)
```

```python
df['CustomerID'].isna().sum()
df['CustomerID'].fillna(0,inplace=True)
```

```python
df['CustomerID']=  df['CustomerID'].astype(int)
```

```python
df['InvoiceDate']= pd.to_datetime(df['InvoiceDate'])
```

```python
df['InvoiceDate_year']= df['InvoiceDate'].dt.year
df['InvoiceDate_month']= df['InvoiceDate'].dt.month
df['InvoiceDate_day']= df['InvoiceDate'].dt.day
df['invoice_date_new']= df['InvoiceDate_year'].astype(str) + '-' + df['InvoiceDate_month'].astype(str) + '-' + df['InvoiceDate_day'].astype(str)
```

```python
df['invoice_date_nft']= pd.to_datetime(df['invoice_date_new'],format= '%Y-%m-%d')
```

```python
keep_data_filter =(df['invoice_date_nft'] >='2011-01-01') & (df['invoice_date_nft'] < '2011-07-01')
filter_df = df[keep_data_filter].reset_index(drop=True)
```
